```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
edges = [(1, 2), (1, 6), (2, 3), (2, 4), (2, 6),  
         (3, 4), (3, 5), (4, 8), (4, 9), (6, 7)] 
  
G.add_edges_from(edges) 
nx.draw_networkx(G, with_label = True) 
```

```python
list(G.neighbors(2))
```

```python
G = nx.DiGraph()
G.add_node(1)
G.add_nodes_from(range(2,10))
G.add_edges_from([(1,2),(1,3),(2,4),(2,6),(2,7)])
nx.draw_networkx(G, with_label = True) 
G.nodes(data=True)
G.edges(data=True)
```

```python
# iterate over all the nodes
# get the out edges of the node
# compute the percentage for the outbound node randomly
for n in list(G.nodes()):
    succ = list(G.successors(n))
    k=np.random.rand(len(succ))
    perc = k / sum(k)
    print(f"hello {n} - number of edges {len(succ)} - Percentage {perc}")
    update_dict = {}
    if (len(succ) > 0):
        for ind, s in enumerate(succ):
            print(f'successor - {s} - {ind} - {G[s]}')
            update_dict[s] = {"value":perc[ind]}
        print(update_dict)
        nx.set_node_attributes(G,update_dict)
    
```

```python
G.nodes(data=True)
```

```python
G1 = nx.DiGraph()
G1.add_nodes_from([("Payments - Web",{"value":1000})])
G1.add_node("Web - Bill Payment")
G1.add_edges_from([("Payments - Web","Web - Bill Payment",{"count":1000})])
G1.add_node("Web - 3rd Party Payments")
G1.add_node("Web - Card Bill Pay")
G1.add_node("Web - Payment History")
G1.add_edges_from([("Web - Bill Payment","Web - 3rd Party Payments",{"count":0}),
                   ("Web - Bill Payment","Web - Card Bill Pay",{"count":0}),
                   ("Web - Bill Payment","Web - Payment History",{"count":0})])



nx.draw_networkx(G1, with_label = True) 
```

```python
G1.nodes(data=True)
```

```python
for n, node_attributes in list(G1.nodes(data=True)):
    node_value = node_attributes["value"]
    succ = list(G1.successors(n))
    k=np.random.rand(len(succ))
    perc = k / sum(k)
    update_edges_dict = []
    update_nodes_dict = {}
    if (len(succ) > 0):
        for ind, s in enumerate(succ):
            print(f'successor - {s} - {ind} - {G1[s]}')
            update_edges_dict.append((n,s,{"perc":perc[ind],"count":perc[ind] * node_value})) 
            update_nodes_dict[s] = {"value":perc[ind] * node_value}
        print(update_dict)
        nx.set_node_attributes(G1,update_nodes_dict)
        nx.set_edge_attributes(G1,update_edges_dict)
```
